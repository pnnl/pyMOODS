## Multi-Criteria Decision-Making - Objective Weighing

```python
import numpy as np

import pymoo
from pymoo.problems import get_problem

import pyDecision as pyd
from pyDecision.algorithm import entropy_method
from pyDecision.algorithm import topsis_method
```

```python
from vis import Loader
```

```python
data_loader = Loader(from_problem="dtlz2")
```

```python
# Load Criterion Type: 'max' or 'min'
criterion_type = ['max', 'max', 'max']

# Dataset
dataset = data_loader.df[["f0", "f1", "f2"]].reset_index(drop=True)

# Call Entropy Function
weights = entropy_method(dataset, criterion_type)
```

```python
# Weigths
for i in range(0, weights.shape[0]):
  print('w(g'+str(i+1)+'): ', round(weights[i], 3))
```

```python
# TOPSIS

# Weights
weights =  [0.25, 0.5, 0.25]

# Load Criterion Type: 'max' or 'min'
criterion_type = ['max', 'max', 'max']

# Call TOPSIS
relative_closeness = topsis_method(dataset, weights, criterion_type, graph=False, verbose=False)
```

```python
np.sort(relative_closeness)[::-1]
```

```python
max_index = np.argsort(relative_closeness)[-1]
data_loader.df.iloc[max_index]
```
