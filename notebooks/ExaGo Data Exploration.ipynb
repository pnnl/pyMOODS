```python
mpc = oc.output_loadloss_at_5_load_scalex_1()
# mpc
```

```python
bus_cols_str = "bus_i	type	Pd	Qd	Gs	Bs	area	Vm	Va	baseKV	zone	Vmax	Vmin	mult_Pmis	mult_Qmis	Pslack	Qslack"
gen_cols_str = "bus	Pg	Qg	Qmax	Qmin	Vg	mBase	status	Pmax	Pmin	Pc1	Pc2	Qc1min	Qc1max	Qc2min	Qc2max	ramp_agc	ramp_10	ramp_30	ramp_q	apf	pgs"
branch_cols_str = "fbus	tbus	r	x	b	rateA	rateB	rateC	ratio	angle	status	angmin	angmax	Pf	Qf	Pt	Qt"
gen_cost_str = "2	startup	shutdown	n	c2	c1	c0"
load_cost_str = "maxallowedloadshed	loadshedcost"
summary_stats_str = "Nbus	Ngen	NgenON	Nline	NlineON	Nload	GenPCap	GenTotalP	GenTotalQ	GenPCapON	LoadTotP	LoadTotalQ	LoadShedP	LoadShedQ"
```

```python
col_names = {}
col_names["bus"] = bus_cols_str.split("\t")
col_names["gen"] = gen_cols_str.split("\t")
col_names["branch"] = branch_cols_str.split("\t")
col_names["gencost"] = gen_cost_str.split("\t")
col_names["loadcost"] = load_cost_str.split("\t")
col_names["summary_stats"] = summary_stats_str.split("\t")
```

```python
for k in col_names.keys():
    print(k)
    mpc[k] = pd.DataFrame(data=mpc[k], columns=col_names[k])
```
